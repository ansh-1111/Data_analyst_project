```python
# Determine the shape of the DataFrame.
print("Shape of the DataFrame:", df.shape)

# Examine the data types of each column.
print("\nData Types:\n", df.dtypes)

# Generate descriptive statistics for numerical columns.
print("\nDescriptive Statistics:\n", df.describe())

# Check for missing values.
print("\nMissing Values:\n", df.isnull().sum())

# Examine unique values in categorical columns.
categorical_cols = ['Country', 'Year', 'Attack Type', 'Target Industry', 'Attack Source', 'Security Vulnerability Type', 'Defense Mechanism Used']
for col in categorical_cols:
    print(f"\nUnique values in '{col}':\n{df[col].unique()[:10]}...\nValue counts:\n{df[col].value_counts()[:5]}")

# Assess the distribution of key columns.
print("\nDistribution of 'Year':\n", df['Year'].value_counts())
print("\nDistribution of 'Attack Type':\n", df['Attack Type'].value_counts()[:10])
print("\nDistribution of 'Target Industry':\n", df['Target Industry'].value_counts()[:10])
```

## Data analysis

### Subtask:
Analyze the distribution of key variables over time and identify trends and patterns.


**Reasoning**:
Analyze the trends in the frequency of different attack types over the years, investigate the relationship between the year and the financial loss, examine the distribution of attacks across different target industries, and explore the correlation between the number of affected users, attack type, and year.  Create visualizations to illustrate these trends and patterns.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Trend of attack types over time
plt.figure(figsize=(12, 6))
sns.countplot(x='Year', hue='Attack Type', data=df)
plt.title('Trend of Attack Types Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Relationship between year and financial loss
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Financial Loss (in Million $)', data=df)
plt.title('Financial Loss Over Time')
plt.xlabel('Year')
plt.ylabel('Financial Loss (in Million $)')
plt.show()


# Distribution of attacks across industries over time
plt.figure(figsize=(12, 6))
sns.countplot(x='Year', hue='Target Industry', data=df)
plt.title('Distribution of Attacks Across Industries Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Correlation between affected users, attack type, and year
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Year', y='Number of Affected Users', hue='Attack Type', data=df)
plt.title('Number of Affected Users by Attack Type and Year')
plt.xlabel('Year')
plt.ylabel('Number of Affected Users')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analyze trends in attack types over time
attack_type_counts = df.groupby(['Year', 'Attack Type']).size().unstack(fill_value=0)
attack_type_counts.plot(kind='line', figsize=(12, 6))
plt.title('Trend of Attack Types Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.legend(title='Attack Type')
plt.tight_layout()
plt.show()

# Analyze the relationship between year and financial loss
plt.figure(figsize=(10, 6))
sns.regplot(x='Year', y='Financial Loss (in Million $)', data=df)
plt.title('Financial Loss over Time')
plt.xlabel('Year')
plt.ylabel('Financial Loss (in Million $)')
plt.show()

# Analyze distribution of attacks across target industries
industry_counts = df.groupby(['Year', 'Target Industry']).size().unstack(fill_value=0)
industry_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Distribution of Attacks Across Industries Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.legend(title='Target Industry')
plt.tight_layout()
plt.show()

# Analyze the correlation between affected users, attack type, and year
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='Number of Affected Users', hue='Attack Type', data=df)
plt.title('Number of Affected Users vs. Year')
plt.xlabel('Year')
plt.ylabel('Number of Affected Users')
plt.show()
```

## Data visualization

### Subtask:
Visualize the key trends and patterns identified in the data analysis.


**Reasoning**:
Visualize the key trends and patterns in the data using appropriate chart types. Focus on threat frequency over time, geographical distribution, relationships between threat categories, financial losses, and affected users.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Threat Frequency Over Time
plt.figure(figsize=(14, 6))
sns.countplot(x='Year', hue='Attack Type', data=df)
plt.title('Threat Frequency Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.legend(title='Attack Type')
plt.tight_layout()
plt.show()

# Geographical Distribution of Threats (Top 10 Countries)
top_10_countries = df['Country'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
top_10_countries.plot(kind='bar')
plt.title('Top 10 Countries Affected by Cyberattacks')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Financial Losses Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Financial Loss (in Million $)', data=df)
plt.title('Financial Losses Over Time')
plt.xlabel('Year')
plt.ylabel('Financial Loss (in Million $)')
plt.show()

# Number of Affected Users Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Number of Affected Users', data=df)
plt.title('Number of Affected Users Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Affected Users')
plt.show()

# Relationship between Threat Categories (Example: Attack Type and Target Industry)
plt.figure(figsize=(12, 6))
sns.countplot(x='Attack Type', hue='Target Industry', data=df)
plt.title('Relationship between Attack Type and Target Industry')
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.legend(title='Target Industry')
plt.tight_layout()
plt.show()
```
